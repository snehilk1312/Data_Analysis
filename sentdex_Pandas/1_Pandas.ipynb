```python
new_df = pd.read_csv('/home/moritz/Desktop/kaggle/avocado-prices/avocado.csv')

new_df = new_df.copy()[new_df['type'] == 'organic']
new_df['Date'] = pd.to_datetime(new_df['Date'])
new_df.sort_values(by='Date', ascending=True, inplace=True)

```

```python
print(new_df.head())
```

```python
graph_df = pd.DataFrame()
for region in new_df['region'].unique():
    print(region)
    region_df = new_df.copy()[new_df['region']==region]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f'{region}_price25ma']]
        
    else:
        graph_df = graph_df.join(region_df[f'{region}_price25ma'])

    
```

```python
print(graph_df.dropna().head())
```

```python
graph_df.plot(figsize=(18,5), legend=False)
```

```python
graph_df.dropna().plot(figsize=(18,5),legend=False)
```
