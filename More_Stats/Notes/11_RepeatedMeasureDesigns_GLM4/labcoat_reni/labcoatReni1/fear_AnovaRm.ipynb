```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pingouin as pg
import scipy.stats as st
import statsmodels.api as sm
import statsmodels.formula.api as smf
%matplotlib inline
```

```
data = pd.read_csv('Field.csv')
```

```
data.head()
```

```
df = pd.melt(data, id_vars=['code'], value_vars=['bhvneg','bhvpos','bhvnone'])
```

```
df.sort_values(['code', 'variable'], inplace=True)
```

```
df.head()
```

```
_ = sns.barplot(x='variable', y='value',data=df)
```

```
df.groupby('variable').describe()['value']
```

```
_ = sns.boxplot(x='variable', y='value',data=df)
```

```
pg.rm_anova(data,effsize='n2')
```

#### shapiro-wilk test

```
st.shapiro(df[df['variable']=='bhvneg']['value'])
```

```
st.shapiro(df[df['variable']=='bhvpos']['value'])
```

```
st.shapiro(df[df['variable']=='bhvnone']['value'])
```

####        The resulting output above indicates that the data are very heavily non‐normal asthe p‐values are all highly significant.



```
pg.qqplot(df[df['variable']=='bhvneg']['value'])
```

```
pg.qqplot(df[df['variable']=='bhvpos']['value'])
```

```
pg.qqplot(df[df['variable']=='bhvnone']['value'])
```

#### **log transforming the scores**

```
df['logval'] = np.log(df['value']+1)
```

```
st.shapiro(df[df['variable']=='bhvneg']['logval'])
```

```
st.shapiro(df[df['variable']=='bhvpos']['logval'])
```

```
st.shapiro(df[df['variable']=='bhvnone']['logval'])
```

```
%%R
fieldData<-read.delim("Field.dat", header =TRUE)
```

```
%%R
fieldData2<-fieldData[, -c(1)]
```

```
%%R
rmanova(fieldData2)
```

#### In the output above we are given a test statistic , F ,  for the effect of information($test) , the degrees of freedom ($df) , the p‐value ($siglevel) , the group means ($tmeans) . Given that the significance level is much less than  .05, we can say that there were significant differences in approach times after hearing different types of information ,F (1.24,94.32) = 78.15 , p<.001.

```
%%R
# post hoc test
rmmcp(fieldData2)
```
