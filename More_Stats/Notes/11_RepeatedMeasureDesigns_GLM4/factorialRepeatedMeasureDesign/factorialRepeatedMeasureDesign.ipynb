```
data = pd.read_csv('Attitude.csv')
```

```
data.head()
```

```
df = pd.melt(data, id_vars=['participant'], value_vars=['beerpos','beerneg','beerneut','winepos','wineneg','wineneut','waterpos','waterneg','waterneu'])
```

```
df.sort_values(['variable'],inplace=True)
df.reset_index(drop=True, inplace=True)
```

```
df['drink']=df['variable'].replace({'beerneg':'beer','beerneut':'beer','beerpos':'beer',
                                    'winepos':'wine','wineneut':'wine','wineneg':'wine',
                                    'waterpos':'water','waterneu':'water','waterneg':'water'})
df['atd'] = df['variable'].replace({'beerneg':'neg','beerneut':'neut','beerpos':'pos',
                                    'winepos':'pos','wineneut':'neut','wineneg':'neg',
                                    'waterpos':'pos','waterneu':'neut','waterneg':'neg'})
df.head()
```

```
%%R
attitudeData<-read.delim("Attitude.dat", header = TRUE)
```

```
%%R
longAttitude <-melt(attitudeData, id = "participant", measured = c( "beerpos", "beerneg", "beerneut", 
                                                                   "winepos", "wineneg", "wineneut", "waterpos", "waterneg", "waterneut"))
```

```
%%R
names(longAttitude)<-c("participant", "groups", "attitude")
```

```
%%R
longAttitude$drink<-gl(3, 60, labels = c("Beer", "Wine", "Water"))
```

```
%%R
longAttitude$imagery<-gl(3,  20,  180,  labels  =  c("Positive",  "Negative", "Neutral"))
```

```
%%R
AlcoholvsWater<-c(1, 1, -2)
BeervsWine<-c(-1, 1, 0)
contrasts(longAttitude$drink)<-cbind(AlcoholvsWater, BeervsWine)
```

```
%%R
NegativevsOther<-c(1, -2, 1)
PositivevsNeutral<-c(-1, 0, 1)
contrasts(longAttitude$imagery)<-cbind(NegativevsOther, PositivevsNeutral)
```

### Posthoc Test

#### **Python Result**

```
from statsmodels.sandbox.stats.multicomp import MultiComparison
multicomp = MultiComparison(df['value'], df['variable'])   # testfunc
```

```
# Bonferroni
com = multicomp.allpairtest(st.ttest_rel, method='bonf')
print(com[0])
```

#### **R Result**

```
%%R
pairwise.t.test(longAttitude$attitude, longAttitude$groups, paired = TRUE, p.adjust.method = "bonferroni")
```

### **Factorial repeated-measures designs as a GLM**

```
%%R
baseline<-lme(attitude ~ 1, random = ~1|participant/drink/imagery, data = longAttitude, method = "ML")
```

```
%%R
drinkModel<-update(baseline, .~. + drink)
```

```
%%R
imageryModel<-update(drinkModel, .~. + imagery)
```

```
%%R
attitudeModel<-update(imageryModel, .~. + drink:imagery)
```

```
%%R
anova(baseline, drinkModel, imageryModel, attitudeModel)
```

```
%%R
summary(attitudeModel)
```

### **Effect sizes for factorialrepeated-measures designs**

```
def rcontrast(t,df):
  return np.sqrt(t**2/(t**2+df))
```

```
# from above anova summary, alcohol vs. water
rcontrast(3.18, 38)
```

```
# beer vs. wine
rcontrast(-1.47, 38)
```

```
# For imagery we contasted negative vs. other
rcontrast(-9.81, 114)
```

```
# alcohol vs. water with negative vs. other imagery
rcontrast(0.69, 114)
```

```
# beer vs. wine with negative vs. other imagery
rcontrast(6.77, 114)
```

```
# alcohol vs. water with positive vs. neutral imagery
rcontrast(0.93, 114)
```

```
# beer vs. wine with positive vs. neutral imagery
rcontrast(-0.80, 114)
```
